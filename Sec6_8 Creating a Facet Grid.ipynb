```python
f,axes=plt.subplots(1,2,figsize=(12,6),sharex=True,sharey=True)
k1=sns.kdeplot(movies.BudgetMillions,movies.AudienceRatings,ax=axes[0])
k2=sns.kdeplot(movies.BudgetMillions,movies.CriticRatings,ax=axes[1])
k1.set(xlim=(-20,160))
#sharex=True will take the same limit of x axis in each subplot
#same for sharey is true.
#(xlim=(-20,160)) will set limit of x axis
```

```python
#CREATING A FACET GRID
```

```python
g=sns.FacetGrid(movies,row='Genre',hue='Genre')
#its not a subplot,we are spilliting up a visulization based on some rules.
#here row is Genre,And we have 7 diffrent types of Genre(Action,Comedy,Horror,...) so we have 7 rows in total.
#Similar for column,if col(column)='Year' than for 2007 to 2011 total 5 column will be added.
```

```python
g=sns.FacetGrid(movies,row='Genre',col='Year')
#click at rectangle at most left side of [0,:] plo tso that all chart will be in 1 page.
```

```python
#example
plt.scatter(movies.CriticRatings,movies.AudienceRatings)
#now we want to map this scatterplot into our FacetGrid
```

```python
g.map(plt.scatter,'CriticRatings','AudienceRatings')
#if you run it alone than it gives you an error because FacetGrid is still not define in this cell.
```

```python
g=sns.FacetGrid(movies,row='Genre',col='Year',hue='Genre')
g=g.map(plt.scatter,'CriticRatings','AudienceRatings')
#so now FacetGrid is defined in this cell so it will run.
#g.map(TYPE OF PLOT,X axis data,Y axis data) is the syntax.

#so noww look at chart at In [56],its difficult to analyse that chart,now we have also spllited it year and Genre wise by FacetGrid.
#So, now its comparitivly easy to analyse.
```

```python
g=sns.FacetGrid(movies,row='Genre',col='Year',hue='Genre')
kws=dict(s=50,linewidth=1,edgecolor='Black')
g=g.map(plt.scatter,'CriticRatings','AudienceRatings',**kws)
#so kws is a disctonary which includes size,boundry width and color of boundry of the dots,which further passed in g.map with **kws syntax.

#Conclusions-Drama charts are getting more and more condense along with years,
#Action movies gets seperated into 2 groups iyear 2011,etc..
```
